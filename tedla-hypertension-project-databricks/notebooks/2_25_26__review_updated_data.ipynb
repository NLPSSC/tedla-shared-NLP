```python
import os
from IPython.core.display import Markdown


def write_message(message: str):
    display(Markdown(message))


root_path = "/var/nfs_share/workspaces/ciphi/westerd/tedla/tedla"

subfolder_paths: list[str] = [
    os.path.join(root_path, name)
    for name in os.listdir(root_path)
    if os.path.isdir(os.path.join(root_path, name))
]

write_message(f"Found {len(subfolder_paths)} subfolders in {root_path}.")

write_message("\n".join([f"- {subfolder_path}" for subfolder_path in subfolder_paths]))
```

```python
empty_parquet_files = []
actual_parquet_files = []

for subfolder in subfolder_paths:

    # find all .parquet files in the subfolder
    parquet_files = [
        os.path.join(subfolder, name)
        for name in os.listdir(subfolder)
        if name.endswith(".parquet")
    ]

    # Check to make sure each parquet file is not empty
    for parquet_file in parquet_files:
        if os.path.getsize(parquet_file) == 0:
            empty_parquet_files.append(parquet_file)
        else:
            actual_parquet_files.append(parquet_file)

if len(empty_parquet_files) == 0:
    write_message("No empty .parquet files found.")
else:
    write_message(f"Found {len(empty_parquet_files)} empty .parquet files:")
    write_message("\n".join([f"- {empty_parquet_file}" for empty_parquet_file in empty_parquet_files]))
```

```python
# Get the total number of rows for each actual parquet file
from pathlib import Path


num_rows_per_dataset = {}

for p in actual_parquet_files:
    # Use pyarrow to read the parquet file and get the number of rows
    import pyarrow.parquet as pq

    write_message(f"Reading {p}...")

    dataset_name = Path(p).parent.name
    if dataset_name not in num_rows_per_dataset:
        num_rows_per_dataset[dataset_name] = 0
    table = pq.read_table(p)

    num_rows_per_dataset[dataset_name] += table.num_rows

for key, value in num_rows_per_dataset.items():
    write_message(f"{key}: {value} rows")
```
