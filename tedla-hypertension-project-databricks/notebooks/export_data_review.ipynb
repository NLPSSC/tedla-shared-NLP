```python
from IPython.core.display import Markdown
display(Markdown(",<br/>".join([x for x in df.columns])))
```

```python
# display(Markdown())
min_date, max_date = df['note_date'].min(), df['note_date'].max()
display(Markdown(f"Min date: {min_date}, Max date: {max_date}"))
```

```python
import matplotlib.pyplot as plt

group_cols = [col for col in df.columns if col.endswith('_group')]
group_sums = df[group_cols].sum().sort_values(ascending=False)

# Shorten labels for readability
short_labels = [col.replace('_notes_group', '').replace('_', ' ') for col in group_sums.index]

fig, ax = plt.subplots(figsize=(12, 6))
bars = ax.bar(short_labels, group_sums.values, color='steelblue', edgecolor='black')

ax.set_title('Counts of Notes by Group', fontsize=14)
ax.set_xlabel('Group', fontsize=12)
ax.set_ylabel('Count', fontsize=12)
ax.set_xticklabels(short_labels, rotation=45, ha='right')

for bar in bars:
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
            f'{int(bar.get_height()):,}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```
