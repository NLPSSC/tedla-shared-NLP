```python
from pathlib import Path

note_to_groups_map_path = Path("/var/nfs_share/workspaces/ciphi/westerd/tedla/tedla/note_id_to_group_membership/note_to_groups_map.parquet")
```

```python
import polars as pl
from IPython.core.display import Markdown

# Read the Parquet file
note_to_groups_map: pl.DataFrame = pl.read_parquet(note_to_groups_map_path)

note_groups: list[str] = [x for x in note_to_groups_map.columns if x != "note_id"]

sample_df: pd.DataFrame = note_to_groups_map.head(10).to_pandas()
display(Markdown("\n\n".join([f"{x}: str" for x in sample_df.columns])))
display(sample_df)
```

```python
from itertools import combinations
from IPython.core.display import Markdown

# Generate all unique, ordered combinations of note_groups (length 2 to len(note_groups))
all_combinations = []
for r in range(1, len(note_groups) + 1):
    all_combinations.extend(combinations(note_groups, r))

# all_combinations is a list of tuples
display(f"Total combinations: {len(all_combinations)}")
columns_to_check = [", ".join([f'"{x}"' for x in comb]) for comb in all_combinations]
counts_rows = []
for c in columns_to_check:
    cols = [col.strip('"') for col in c.split(", ")]
    count = note_to_groups_map.filter(pl.all_horizontal([pl.col(col) for col in cols])).height
    counts_rows.append((", ".join([x for x in cols]), count))
```

```python
import pandas as pd
from IPython.core.display import Markdown

# Create a DataFrame from counts_rows
counts_df = pd.DataFrame(counts_rows, columns=["Note Groups", "Notes in Groups"])

# Filter and display rows where count > 0
display(Markdown(counts_df[counts_df["Notes in Groups"] > 0].to_markdown(index=False)))

total_notes = counts_df["Notes in Groups"].sum()
display(Markdown(f"**Total Notes in Groups:** {total_notes:,}"))
```
